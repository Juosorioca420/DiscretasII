```python
import numpy as np
#import itertools as it
```

```python
def generador(n): #Arreglo de permutaciones de pasos iguales uno a uno hasta completar un ciclo de rotaciones 360/n (conservar adyacencia)
  P = []
  neutro = list( range(1, n+1) ) #permutaciones de rotación
  reflexion = sorted(neutro, reverse = True) #permutaciones de reflexión

  for i in range(n):
    pn = neutro[i: n] + neutro[:i]
    pi = reflexion[i: n] + reflexion[:i]

    P.insert(i, pn); P.insert(i + n, pi)
  
  return P

def cayley(P):
  n = len(P); m = len(P[0])
  cayley = []
  for i in range(n):
    fila = []
    for j in range(n):
      compuesta = [ P[i][ P[j][k] - 1 ] for k in range(m) ]
      index = P.index(compuesta) 
      fila.append( f'r{index}' ) if index < 6 else fila.append( f's{index - 6}' )

    cayley.append(fila)
    
  return cayley
```

```python
def main(n = 6):
  P = generador(n)
  for i in range( len(P) ): 
   print( f'r{i} : {P[i]}' ) if i < 6 else print( f's{i - 6} : {P[i]}' )

  print()

  C = cayley(P); print( f'Tabla de Cayley.\n' )
  C = np.array(C); print(C)
  
  return C, P

C, P = main()
```
